### 时间序列的特点
时间序列问题不同于一般的回归预测问题，因为时间序列具有四种特征：
1. 长期趋势(trend, T) 较长一段时间内发展的趋势
2. 周期性(Seasonal, S) 一定时间为期间所呈现的周期性连续波动
3. 季节性(Cyclical, C) 由于季节的变化而引起的有规律的波动
4. 随机性(Irregular, I) 噪声对时间序列造成的影响

### 时间序列模型

#### 加法模型

$Y_{t}=T_{t}+S_{t}+C_{t}+I_{t}$

加法模型四种成分相互独立，某种成分的变动并不影响其他成分变动。各个成分都用绝对量表示，并且具有相同的量纲。

#### 乘法模型

$Y_{t}=T_{t} \times S_{t} \times C_{t} \times I_{t}$

乘法模型四种成分保持相互依存的关系，长期趋势用绝对量表示，具有和时间序列本身相同的量纲，其他成分则用相对量表示。

### ARIMA
ARIMA(Autoregressive Integrated Moving Average Model，自回归积分滑动平均)针对非平稳时间序列所建立的模型。

ARIMA相比于自回归移动平均过程ARMA(p,q)多了差分的过程，即当时间序列存在趋势性或周期性时通过差分将非平稳数据转化为平稳数据。K阶差分就是将相距为K的两个序列值相减。若一个时间序列可以通过差分转化为平稳数据，则可以通过ARIMA模型进行分析。

那什么是平稳呢？

就是时间序列围绕常数上下波动且波动范围有限，即有常数均值和常数方差；若存在明显的趋势或周期性，则它们的均值和方差随着时间的变化不断变化，那么它通常不是平稳序列。

如何确定时间序列是否是平稳呢？
1. 直接画出趋势图，看趋势进行判断
2. 根据自相关图和偏相关图，平稳序列的自相关图和偏相关图，要么拖尾，要么是截尾
3. 单位根检验，存在单位根就是非平稳序列

#### AR
AR模型（Autoregressive Model，自回归模型），利用前期数据点和后期数据点的相关关系（又称为自相关），建立前期数据为X值，后期数据为Y值的线性回归模型。

<img src="./imgs/ts02.png" width = "350" height = "350" />

如下图所示
$$
\begin{bmatrix} x_{1} & ... & x_{k} \\ x_{2} & ... & x_{k+1} \\ x_{3} & ... & x_{k+2} \\  x_{n-k} & ... & x_{n-1} \end{bmatrix} \space \begin{bmatrix} θ_{0} \\ θ_{1} \\ . \\  .\\ . \end{bmatrix} = \begin{bmatrix} x_{k+1} \\ x_{k+1} \\ x_{k+3} \\ x_{n} \end{bmatrix}
$$

解释下，白噪声类似线性回归中的误差，这里可以将白噪声理解成时间序列数值的随机波动，这些波动的总和为0。

#### MA
MA模型（Moving Average，滑动平均模型），某个时间点的数值等于白噪声的加权和。与AR类似，只不过之前是通过不同阶滞后序列来拟合y，而这里使用不同阶滞后的白噪声来拟合y。
