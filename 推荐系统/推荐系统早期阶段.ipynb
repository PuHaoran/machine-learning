### 基础知识

#### 什么是推荐系统？
1.为什么需要推荐系统？

万物互联是大趋势，而人是建立连接的关键，而推荐系统就是根据已有连接来预测未知连接的。

2.什么是推荐系统？

根据已有连接，预测未来的连接。

3.你是否真的需要推荐系统？

1）物品足够

2）用户和物品连接数量足够

3）可通过如下指标判断新增长连接是否主要靠新用户和新物品贡献

$\frac{\triangle{connection}}{\triangle{user}\times\triangle{item}}$（分子是新增的连接数，分母是新增活跃用户和新增物品）

#### 推荐系统经典问题

1.显式反馈和隐式反馈

推荐系统问题模式包括：

1）评分预测 

    用户消费完物品之后给出一个打分（1~5），我们根据已有用户打分物品来预测其他物品打分

2）行为预测

    1.直接预测用户行为本身（例如购买、点击、收藏）发生的概率，即CTR预估。
    2.预测物品的相对排序

隐式反馈更加受重视的原因：

1）数据相比评分数据更加稠密

2）隐式反馈常常和模型的目标函数关联更加密切，更容易和AB测试指标挂钩

3）行为预测解决的是80%的问题，而评分预测解决的是那最后20%的问题

2.冷启动问题

冷启动面对对象是新用户或不活跃用户，新物品或展示次数较少物品，对待冷启动通常的解决方式是给它加热，然后想办法引入数据，从已有数据中主动学习（一种半监督学习）

3.探索与利用

推荐系统中一般采用给用户大部分推荐他感兴趣的，小部分探索新的兴趣。类似一边收割长好的韭菜，一边播种新韭菜。

### 内容推荐

#### 什么是用户画像？

1.为什么需要用户画像？

大型系统一般分为召回和排序两个阶段，召回阶段会对全量物品做粗排，筛选出一部分物品，降低排序阶段的计算量。而用户画像除了会用在召回阶段还会用在排序阶段。

对于一个早期的推荐系统，基于内容的推荐离不开为用户构建一个初级的用户画像。

2.构建内容画像的关键

1）维度

    1.用户和物品的维度要相互理解
    2.维度和维度数量都是拍脑袋决定的
    
2）量化
    
    不要为了用户画像而用户画像，它只是推荐系统的一个副产品；一般在生产系统中，会根据推荐系统的好坏（排序好坏、召回覆盖等指标）反向优化用户画像
    
3.构建用户画像的方法
  
1）查户口

    直接使用原始数据作为用户画像的内容，如注册资料等人口统计学信息，或购买历史，阅读历史，这通常对于冷启动等场景非常有用

2）堆数据

    堆积历史数据，做统计。常见的兴趣标签，就是这一类，从历史数据中挖掘出标签，然后再标签维度上做数据统计，用统计结果做量化结果。

3）黑盒子
    
    采用机器学习的方法学习出人类木法理解的稠密向量
    1.隐语义模型构建用户阅读兴趣
    2.矩阵分解得到隐因子
    3.使用深度学习模型学习用户的Embedding向量

#### 从文本到用户画像

对于一个早期的推荐系统来说，基于内容的推荐离不开用户构建一个初级的画像

用户画像包括：

1）用户端

    注册信息中的姓名、个人签名
    评论、动态、日记、搜索信息
    私信

2）物品端

    物品的描述、标题
    物品本身的内容
    物品其他属性的文本
    
如何从文中提取用户画像所需要的特征？

1）把非结构化的文本结构化，去粗取精，保留关键信息
    
    使用成熟的NLP算法可以从文本中得到以下信息
    1.关键词提取
        最基础的标签来源，也为其他文本分析提供基础数据，例如TF-IDF和TextRank
    2.实体识别
        识别出人物、地点、影视、历史事件、热点事件等，常用基于词典结合CRF模型（在应对以实体为代表的序列标注问题上，工业级工具spacy更具效率）
    3.内容分类
        将文本按照某个分类体系分裂，用分类来表达更粗粒度的结构化信息
    4.类簇编号
        在无人指定分类体系情况下，采用无监督方法将文本划分多个类簇也很常见，类簇编号是用户画像的常见构成
    5.主题模型
        从大量文本中学习出主题向量，然后预测新文本在各个主题上的概率分布，主题向量是用户画像的常用构成（在初创公司没有业务专家制定分类体系，若能在文本数据上跑一个LDA模型，那就很美好啦，K值可以通过实验进行对比，每次计算两两主题之间的平均相似度，选择一个较低的K值；得到文本在各主题上的分布，选择概率最大的前几个主题作为文本的主题；LDA在工程上难并行，若数据没达到海量，可以提高单机配置，LDA的库是gensim）
    6.嵌入向量
        word2vec训练词嵌入向量，挖掘语义信息。
        ①计算词与词之间的相似性，扩充结构化标签
        ②累加得到文本的稠密向量
        ③用于聚类，得到比词向量聚类更好的语义聚类效果
 
2）把物品结构化数据，传递给用户，与用户自己的结构化数据合并
    
    我们将用户端、物品端数据结构化，得到了标签（关键词、分类）、主题、词嵌入向量，那如何把物品的结构化信息给用户呢？
    1.一种简单粗暴的方法是把用户产生过行为的物品标签累积在一起
    2.另一种方法是把用户对物品的行为，有消费没消费看做一个分类问题，那么挑选他实际感兴趣的标签就变成了特征选择问题。（文本特征选择最常用的方法：卡方检验CHI和信息增益IG）
        ①把物品的结构化内容看做文档；
        ②把用户对物品的行为看做是类别；
        ③每个用户看见过的物品就是一个文本集合；
        ④在这个文本集合上使用特征选择算法选出每个用户关心的东西。
        （PS：这些方法在离线阶段批量完成，把用户的画像生成配置成离线任务，每天更新一遍，次日就可以使用新的画像了。）
   
#### 做好基于内容的推荐系统

1.为什么需要搭建基于内容的推荐系统？

基于内容的推荐系统一般是推荐系统的起步阶段，而且会持续存在，重要性不可替代。

1）内容数据尤其是文本，只有深入挖掘，就可以挖掘出有用的信息供推荐系统使用

2）产品冷启动阶段，没有用户行为，别无选择。而基于内容的系统可以把新物品推荐出去

2.如何做基于内容的推荐系统？

1）抓 抓取数据补充内容源，增加分析维度

2）洗 洗除冗余内容、垃圾内容、政治色情等敏感内容

3）挖 对内容做更深入挖掘

4）算 匹配用户的兴趣和物品属性，计算合理的相关性

3.内容分析的产出？

内容分析的产出包括：

1）结构化的内容库 

    用途：结合用户反馈去学习用户画像

2）内容分析模型

    用途：当新物品进入，需要实时地推荐出去，这时候对内容实时分析，提取结构化内容，然后再用于用户画像匹配。
    
    模型：1.分类器模型2.主题模型3.实体识别模型4.嵌入模型
    
4.内容推荐算法

1）最简单的推荐算法是直接计算相似性，用户画像表示为稀疏的向量，内容端也有稀疏向量，两者之间计算余弦相似度，根据相似度对推荐物品排序（对内容分析深入的话，效果还是很不错的，而且基于内容的推荐天然有一个优点，可解释性强）
    
2）另一种是训练预估模型。一般我们为了提高某种行为的转化率，如点击、收藏、转发。标准做法是收集这类行为的日志数据，转化为训练样本，训练预估模型。（详细介绍下，每一条样本由两部分组成，一部分是特征，包含用户端的画像内容，物品端的结构化内容，可选的还有日志记录时的一些上下文场景信息，如时间、地理位置、设备等，另一部分是用户行为，作为标注信息，包含有反馈和无反馈两种。用这样的样本训练一个二分类，常用LR或梯度提升树或二者结合，在推荐匹配预估用户行为发生的概率，按照概率排序，这个方法科学合理，可以一直迭代优化。）

### 总结
本篇简单描述了什么是推荐系统，提出了推荐系统中的几个经典问题，然后又阐述了算法工程师应该具备的思维模式。重点介绍了推荐系统中用户画像的构建方法，以及在推荐系统早期阶段，如何建立一个内容推荐系统。文章给出了一个清晰的内容推荐系统框架：对内容推荐系统而言，内容分析是关键，推荐算法可以采用快糙猛的相似度计算，也可以采用机器学习思路的预估模型，在拥有大量用户行为后也可以采用机器学习思路训练预估模型。

关键词：用户画像，内容分析，内容推荐系统。
